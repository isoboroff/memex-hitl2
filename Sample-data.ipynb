```r
library(ggplot2)
library(dplyr)
library(readxl)
library(tidyr)
library(forcats)
```

```r
df <- read_excel("Sample_Question_Scoring.xlsx")
```

```r
df
```

```r
data <- tail(df, -1) %>% 
    gather(key="question", value="correct_found", -PARTICIPANT) %>% 
    mutate(question = fct_recode(question, "1" = "QUESTION 1", "2" = "QUESTION 2","3" = "QUESTION 3", "4" = "QUESTION 4"))
data
```

```r
gtruth <- data %>% filter(PARTICIPANT == 'GROUND TRUTH') %>% select(question, correct_found)
gtruth
```

```r
data2 <- data %>% filter(PARTICIPANT != 'GROUND TRUTH') %>% 
    left_join(gtruth, by="question") %>%
    rename(found = correct_found.x, known_correct = correct_found.y) %>%
    mutate(recall = found / known_correct) %>%
    mutate(participant = fct_reorder(PARTICIPANT, as.integer(PARTICIPANT)))
data2
```

```r
options(jupyter.plot_mimetypes = 'image/png')
```

```r
print(ggplot(data2, aes(participant, recall)) + geom_boxplot())
```

```r
ggplot(data2, aes(question, recall)) + geom_boxplot()
```
